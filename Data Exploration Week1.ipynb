3. Can market consolidation be estimated using estimated sales data and a measure of market concentration like the HerfindahlHirschman Index (HHI)?

```python
#Find top ten parent companies
grocery_t10_22 = grocery_sic.groupby('PARENT NUMBER', dropna=False).size().reset_index(name='COUNT')
grocery_t10_22.sort_values('COUNT', ascending=False, inplace=True)
# Create other row
other_row = grocery_t10_22.iloc[10:]
other_column = other_row['COUNT'].sum()
# Remove other rows
grocery_t10_22 = grocery_t10_22.iloc[:10]
grocery_t10_22.loc[len(grocery_t10_22)] = ["Other", other_column]
grocery_t10_22['PARENT NUMBER'] = grocery_t10_22['PARENT NUMBER'].astype(str)
grocery_t10_22['Percent'] = grocery_t10_22['COUNT'] / len(grocery_sic) * 100
grocery_t10_22

#Percent tells us MarketShare (basically how much space it takes up in the grocery industry)
#Count tells us how many grocery stores are under a specific parent company/ number 
#where "NAN" means there's no data available and "Other" is everything from 11 onwards
```

```python
# Every single grocery store included in this dataframe (not just top 10 )
grocery_22 = grocery_sic.groupby('PARENT NUMBER', dropna=False).size().reset_index(name='COUNT')
grocery_22.sort_values('COUNT', ascending=False, inplace=True)

grocery_22['PARENT NUMBER'] = grocery_22['PARENT NUMBER'].astype(str)
grocery_22['Percent'] = grocery_22['COUNT'] / len(grocery_sic) * 100
grocery_22 = grocery_22.reset_index(drop=True)
grocery_22
```

**The HHI is a commonly accepted measure of market concentration. It is calculated by squaring the market share of each firm competing in a market and then summing the resulting numbers. It can range from close to 0 to 10,000. The U.S. Department of Justice uses the HHI for evaluating potential merger antitrust issues.**

*Source: Investopedia*

```python
#Calculating the HHI for the grocery industry in 2022 in the US

import numpy as np
#function used to calculate hhi 

def hhi(num):
    return np.square(num).sum()    

# hhi_df = grocery_22.groupby('PARENT NUMBER').agg({'Percent': hhi})

hhi(grocery_22['Percent'][1:]) #without the NAN PARENT NUMBER( independent stores that take up about 49% of the market)
```

```python
#Slicing to get the top 10 
hhi(grocery_22['Percent'][1:11])
```

A market with an HHI **below 1,500** is considered a **competitive marketplace**, an HHI of 1,500 to 2,500 is moderately concentrated, and an HHI of 2,500 or greater is highly concentrated. 

As a general rule, mergers that increase the HHI by more than 200 points in highly concentrated markets raise antitrust concerns, as they are assumed to enhance market power.
