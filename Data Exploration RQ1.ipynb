```python
#importing needed packages 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
#reading in the data 

data = pd.read_csv('/srv/data/my_shared_data_folder/rafi/2022_Business_Academic_QCQ_grocery.csv')
data.head()
```

```python
#get all columns to show 

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

data.head()
```

```python
#checking the number of parent companies 

data["PARENT NUMBER"].value_counts()
```

```python
#checking the number of grocery stores 

data["PRIMARY SIC CODE"].value_counts()
```

```python
#GROCERY_SIC = "5411"

#Locating the string where 5411 for groceries and matching to parent companies  

data.groupby(['PARENT NUMBER','PRIMARY SIC CODE'])
```

```python
#plotting a bar graph to show count of sales in 2022

data["SALES VOLUME (9) - LOCATION"].value_counts().plot(kind="bar")

plt.title("Count of Sales ")
plt.xlabel("Sales")
plt.ylabel("Count")
plt.show()
```
